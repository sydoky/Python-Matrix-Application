```python
import pandas as pd
import numpy as np
import re 
```

```python
from numpy.core.fromnumeric import transpose
#Build a userinterface

print("***************** Welcome to the Python Matrix Application***********")

while True:
  game = input("Do you want to play the Matrix Game?\nEnter Y for Yes or N for No:")

  if game == "Y":
    while True:
      phone_number = input("Enter your phone number XXX-XXX-XXXX: ")
      if re.match("\d{3}-\d{3}-\d{4}",phone_number):
        break
      else:
        print("Your phone number is not in correct format. Please renter: ")

    while True:
      zip_code = input("Enter your zip code+4 (XXXXX-XXXX): ")
      if re.match("\d{5}-\d{4}", zip_code):
        break
      else:
        print("Your zip code is not in correct format. Please renter: ")

    print("Enter your first 3x3 matrix: ")
    mat1=[]
    for i in range(0,3):
      row=input().split() #split converts string into list 
      row=list(map(int,row)) #map will define position of each element [1,1 1,2, 1,3] next row [2,1 2,2 2,3] and so on
      mat1.append(row) #we are appending each value to the matrix list

    print("Your first matrix is: ")
    for r in range(0,3): #r is for row
      for c in range(0,3): #c is for column
        print(mat1[r][c],end=" ")
      print()

    print("Enter your second 3x3 matrix: ")
    mat2=[]
    for i in range(0,3):
      row=input().split()
      row=list(map(int,row))
      mat2.append(row)

    print("Your second matrix is: ")
    for r in range(0,3):
      for c in range(0,3):
        print(mat2[r][c],end=" ")
      print()

    print("Select a Matrix Operation from the list below: ")

    print("a. Addition \nb. Subtraction\nc. Matrix Multiplication\nd. Element by element multiplication")
    selection = input() 
    if selection == "a":
      mat1 = np.array(mat1)
      mat2 = np.array(mat2)
      addition = mat1 + mat2
      print("You selected Addition. The results are: ")
      for r in range(0,3):
        for c in range(0,3):
          print(addition[r][c],end=" ")
        print()
    #next we need to do transpose (converting rows into columns)
      print("The Transpose is: ")
      transpose1 = np.transpose(addition)
      for r in range(0,3):
        for c in range(0,3):
          print(transpose1[r][c], end=" ")
        print() 

      print("The row and column mean values of the results are: ")

      print("Row:",np.mean(addition,axis=1)) #axis 1 is for row
      print("Column:",np.mean(addition,axis=0)) #axis 2 is for column 

    elif selection == "b":
      mat1 = np.array(mat1)
      mat2 = np.array(mat2)
      subtraction = mat1 - mat2
      print("You selected Subtraction. The resulsts are: ")
      for r in range(0,3):
        for c in range(0,3):
          print(subtraction[r][c], end=" ")
        print()
      print("The transpose is: ")
      transpose1 = np.transpose(subtraction)
      for r in range(0,3):
        for c in range(0,3):
          print(transpose1[r][c], end=" ")
        print()

      print("The row and column mean values of the results are: ")

      print("Row:",np.mean(subtraction, axis=1))
      print("Column:", np.mean(subtraction, axis=0))

    elif selection == "c":
      mat1 = np.array(mat1)
      mat2 = np.array(mat2)
      matrix_multiplication = np.matmul(mat1, mat2) 
      print("You selected Matrix Multiplication. The resulsts are: ")
      for r in range(0,3):
        for c in range(0,3):
          print(matrix_multiplication[r][c], end=" ")
        print()
      print("The transpose is: ")
      transpose1 = np.transpose(matrix_multiplication)
      for r in range(0,3):
        for c in range(0,3):
          print(transpose1[r][c], end=" ")
        print()

      print("The row and column mean values of the results are: ")

      print("Row:",np.mean(matrix_multiplication, axis=1))
      print("Column:", np.mean(matrix_multiplication, axis=0))

    elif selection == "d":
      mat1 = np.array(mat1)
      mat2 = np.array(mat2)
      element_by_element_multip = mat1 * mat2 
      print("You selected element by element multiplication. The resulsts are: ")
      for r in range(0,3):
        for c in range(0,3):
          print(element_by_element_multip[r][c], end=" ")
        print()
      print("The transpose is: ")
      transpose1 = np.transpose(element_by_element_multip)
      for r in range(0,3):
        for c in range(0,3):
          print(transpose1[r][c], end=" ")
        print()

      print("The row and column mean values of the results are: ")

      print("Row:",np.mean(element_by_element_multip, axis=1))
      print("Column:", np.mean(element_by_element_multip, axis=0))

  else:
    print("Sorry to see you go")
    break
```
